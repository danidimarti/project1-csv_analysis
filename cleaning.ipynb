```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data/attacks.csv", encoding='cp1252')
```

```python
df.dropna() #drop rows of data that are empty
```

```python
df.isna() #checks if there's still rows with empty cells 

```

```python
df.columns=df.columns.str.lower().str.replace(' ','-')
df.columns= df.columns.str.rstrip(' ')
```

```python
df.sample()
```

```python
lst = list(df.columns)
print(lst)
```

```python
df['species-'] = df['species-'].str.rstrip('-')
```

```python
#fill NaN values from species column with 'Unknown'
df['species-'].fillna('Unknown', inplace=True)
```

```python
#fill NaN values from activity column with 'Unknown'
df['activity'].fillna('Unknown', inplace=True)
```

```python
# Fix capitalization of fatal(y/n) 
df['fatal-(y/n)'] = df['fatal-(y/n)'].str.upper()
```

```python
df['fatal-(y/n)'] = df['fatal-(y/n)'].str.strip(" ")
```

```python
# if column not Y or N, assign as UNKNOWN
df['fatal-(y/n)'] = np.where((df["fatal-(y/n)"] == "2017") | (df["fatal-(y/n)"] == "M"), "UNKNOWN", df["fatal-(y/n)"])
df["fatal-(y/n)"].value_counts()
```

```python
df.rename({'species-': "species", "sex-": "sex"}, axis=1, inplace=True)
```
